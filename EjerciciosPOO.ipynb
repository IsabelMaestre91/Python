**CLASE PERSONA**

```python
import datetime

class Persona:
  def __init__(self, nombre, apellidos, fecha_nacimiento, dni, direccion, localidad, telefono):
    self.nombre = nombre
    self.apellidos = apellidos
    self.fecha_nacimiento = fecha_nacimiento
    self.dni = dni
    self.direccion = direccion
    self.localidad = localidad
    self.telefono = telefono

  def presentarse(self):
    print("Hola soy: ", self.nombre, self.apellidos)
  def direccion_postal(self):
    print("direccion de envio postal: ")
    print(self.direccion, self.localidad, self.telefono)
  def edad(self):
    today = datetime.datetime.today()
    born = datetime.datetime.strptime(self.fecha_nacimiento, '%Y-%m-%d').date()
    edad = today.year - born.year - ((today.month, today.day) < (born.month, born.day))
    print("Edad:", edad)
  def datos(self):
    print("datos de facturación: ")
    print(self.nombre, self.apellidos, self.dni, self.direccion, self.localidad)

persona1 = Persona("Martina", "Perez", "1998-06-23", "20053487M", "C/San Juan N5", "Castellon", "658732149")
persona1.presentarse()
persona1.direccion_postal()
persona1.edad()
persona1.datos()

persona2 = Persona("Valentin", "Barcelo", "2000-05-19", "50153623L", "C/Almeria N36", "Huelva", "645251325")
persona2.presentarse()
persona2.direccion_postal()
persona2.edad()
persona2.datos()

persona3 = Persona("Leonardo", "Ponce", "1977-12-15", "56458971N", "C/Pinto Velazquez N35", "Galicia", "635321489")
persona3.presentarse()
persona3.direccion_postal()
persona3.edad()
persona3.datos()
```

**CLASE CURSO**

```python
class Curso:
  def __init__(self, nombre, horario, fecha_inicio, fecha_finalizacion, horas_lectivas, codigo, profesor):
    self.nombre = nombre
    self.horario = horario
    self.fecha_inicio = fecha_inicio
    self.fecha_finalizacion = fecha_finalizacion
    self.horas_lectivas = horas_lectivas
    self.codigo = codigo
    self.profesor = profesor

class Persona:
  def __init__(self, nombre, apellidos):
    self.nombre = nombre
    self.apellidos = apellidos

una = Persona("Alejandro", "Sanchez")
curso = Curso("Programación de Sistemas Infomáticos", "15:00 a 21:00", "28-01-2024", "31-07-2024", 630, "2510", una)

print(curso.nombre)
print(curso.horario)
print(curso.fecha_inicio)
print(curso.fecha_finalizacion)
print(curso.horas_lectivas)
print(curso.codigo)
print(curso.profesor.nombre)
print(curso.profesor.apellidos)
```
