**WHILE**

```python
num= int(input("Introduzca un número \n"))
multiplicador = 0
print("el número seleccionado es \n", num)

while  num < 0:
    print("el número introducido ",num," no pude ser negativo.")
    num= int(input("Introduzca otro número \n"))
    print("el número ingresado es \n",num)

while multiplicador <= 10:
    resultado = num * multiplicador
    print(num, 'x', multiplicador, '=', resultado)
    multiplicador+=1
```

**ADIVINA EL NÚMERO SIN OPCIONES**

```python
num = 20

intento = int(input("Introduce un número entre 1 y 20:"))

while num != intento:
  intento = int(input("Introduce un número entre 1 y 20:"))

print("¡¡BRAVO!!")

```

**ADIVINA EL NÚMERO CON OPCIONES**

```python
num = 20

intento = int(input("Introduce un número entre 1 y 20:"))
contador = 1

while num != intento and contador <5:
  intento = int(input("Introduce un número entre 1 y 20:"))
  contador = contador + 1

if intento == num:
    print("¡¡BRAVO!!")
else:
  print("¡¡HAS PERDIDO!!")
```

**CAJERO CON MULTIPLES FUNCIONES**

```python
def cajero_automatico():
  saldo = 5000
  numero = True

  while numero:

    print("Bienvenido al cajero automático.")
    print("1.Consulta saldo")
    print("2.Retirar dinero")
    print("3.Ingresar dinero")
    print("0. Salir")

    opción = input("Elija una de las opciones por favor:")

    if numero == "1":
      print(f"Su saldo actual es de:,{saldo}€")

    elif numero == "2":
        retirar = int(input("¿Cuánto dinero desea retirar?"))
        if retirar > saldo:
          print("No hay suficiente saldo.")
        else:
          saldo -= retirar
          print(f"Ha retirado {retirar}. Su saldo actual es {saldo}. ")

    elif numero == "3":
        ingresar=int(input("¿Cuánto dinero desea ingresar?"))
        saldo += ingresar
        print(f"Ha depositado {ingresar} y ahora tiene {saldo} €")

    elif numero == "0":
      numero = False
      print("Gracias por su visita y vuelva pronto")

    else:
      print("Error!! Operación debe ser 1,2, ó 3")

cajero_automatico()
```
