**CLASE APP**

```python
class Animal:
  def __init__(self, nombre, especie, n_patas):
    self.Nombre = nombre
    self.Especie = especie
    self.N_Patas = n_patas

class APP(Animal):
  def __init__(self, nombre, especie, n_patas, nivel_peligrosidad, tipo_peligrosidad):
    super().__init__(nombre, especie, n_patas)
    self.__nivel_peligrosidad = nivel_peligrosidad
    self.__tipo_peligrosidad = tipo_peligrosidad

  def get_nivel_peligrosidad(self):
    return self.__nivel_peligrosidad

  def get_tipo_peligrosidad(self):
    return self.__tipo_peligrosidad

animal = APP("tigre de bengala","Panthera tigris tigris",4,7,"arañazos")
animal_1 = APP("cocodrilo","Crocodylidae",4,9,"mordiscos")

print(f"Hola soy un/a {animal.Nombre}, mi especie es {animal.Especie}, tengo {animal.N_Patas} patas, mi nivel de peligrosidad es {animal.get_nivel_peligrosidad()} y mi tipo de peligrosidad es {animal.get_tipo_peligrosidad()}")
print(f"Hola soy un/a {animal_1.Nombre}, mi especie es {animal_1.Especie}, tengo {animal_1.N_Patas} patas, mi nivel de peligrosidad es {animal_1.get_nivel_peligrosidad()} y mi tipo de peligrosidad es {animal_1.get_tipo_peligrosidad()}")
```

**CALCULADORA CIENTIFICA**

```python
class Calculadora:
  def __init__(self):
    self.__x = 0
    self.__y = 0

  def set_number(self, x,y):
    self.__x = x
    self.__y = y

  def get_x(self):
    return self.__x

  def get_y(self):
    return self.__y

  def sumar(self):
    return self.__x + self.__y

  def restar(self):
    return self.__x - self.__y

  def multiplicacion(self):
    return self.__x * self.__y

  def division(self):
    if self.__y != 0:
      return self.__x / self.__y
    else:
      return "No se puede dividir"

class Cientifica(Calculadora):
  def __init__(self):
    super().__init__()

  def potencia(self):
    exponente = int(input("Introduzca el exponente: "))
    return self.get_x() ** exponente

  def raiz_cuadrada(self):
    return self.get_x() ** 0.5

cientifica = Cientifica()

x = float(input("Introduzca un numero: "))
y = float(input("Introduzca otro numero: "))

cientifica.set_number(x, y)

print("El resultado es: ", cientifica.sumar())
print("El resultado es: ",cientifica.restar())
print("El resultado es: ", cientifica.raiz_cuadrada())
print("El resultado es: ",cientifica.multiplicacion())
print("El resultado es: ",cientifica.division())
print("El resultado es: ",cientifica.potencia())




```

**PERSONAL TRABAJO**

```python
from datetime import datetime

class Persona:
  def __init__(self, nombre, apellidos, dni, edad, fecha_nacimiento):
    self.Nombre = nombre
    self.Apellidos = apellidos
    self.DNI = dni
    self.Edad = edad
    self.Fecha_Nacimiento = fecha_nacimiento

  def edad_persona(self):
    today = datetime.now()
    edad = today.year - self.Fecha_Nacimiento.year - ((today.month, today.day) < (self.Fecha_Nacimiento.month, self.Fecha_Nacimiento.day))
    return edad

class Trabajador(Persona):
  def __init__(self, nombre, apellidos, dni, edad, fecha_nacimiento, horario, fecha_incorporacion, salario, puesto):
    super().__init__(nombre, apellidos, dni, edad, fecha_nacimiento)
    self.__Horario = horario
    self.__Fecha_Incorporacion = fecha_incorporacion
    self.__Salario = salario
    self.__Puesto = puesto

  def antiguedad(self):
    today = datetime.now()
    antiguedad = today - self.__Fecha_Incorporacion
    return antiguedad.days // 365, (antiguedad.days % 365) // 30, (antiguedad.days % 365) % 30

  def get_horario(self):
    return self.__Horario

  def set_horario(self, horario):
    self.__Horario = horario

  def get_fecha_incorporacion(self):
    return self.__Fecha_Incorporacion

  def set_fecha_incorporacion(self, fecha_incorporacion):
    self.__Fecha_Incorporacion = fecha_incorporacion

  def get_salario(self):
    return self.__Salario

  def set_salario(self, salario):
    self.__Salario = salario

  def get_puesto(self):
    return self.__Puesto

  def set_puesto(self, puesto):
    self.__Puesto = puesto

fecha_nacimiento = datetime(2003, 9, 21)
fecha_incorporacion = datetime(2020, 6, 1)
trabajador = Trabajador("Joan", "Castelló", "", 18, fecha_nacimiento, "8:00 - 13:00", fecha_incorporacion, 1200, "gerente")

print("nombre: ", trabajador.Nombre)
print("apellidos: ",trabajador.Apellidos)
print("edad: ", trabajador.edad_persona())
print("horario: ", trabajador.get_horario())
print("fecha de incorporacion: ",trabajador.get_fecha_incorporacion().strftime("%d-%m-%Y"))
print("salario: ", trabajador.get_salario())
print("puesto: ", trabajador.get_puesto())

años,meses,dias = trabajador.antiguedad()
print("antigüedad en la empresa: ",años,"años",meses,"meses", dias,"dias.")
```
