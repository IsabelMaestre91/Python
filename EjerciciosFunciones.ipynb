**ECUACIÓN DE SEGUNDO GRADO**

```python
from math import sqrt

def valor1():
  a = int(input("Introduzca el valor a: "))
  b = int(input("Introduzca el valor b: "))
  c = int(input("Introduzca el valor c: "))
  return a,b,c

def valor2(a,b,c):
  formula = b ** 2 - 4 *a*c
  if formula <0:
    return None, None
  else:
    x1 = (-b + sqrt(formula)) / (2 * a)
    x2 = (-b - sqrt(formula)) / (2 * a)
    return x1,x2

def solucion(x1,x2):
  print("La solución de la ecuación es: ")
  print("x1 =", x1)
  print("x2 =", x2)

a,b,c = valor1()
x1,x2 = valor2(a,b,c)

if x1 is None:
  print("Los números elegidos no son compatibles a la hora de realizar la ecuación.")
else:
  solucion(x1,x2)
```

 **CONVERSOR DE DÍAS DE LA SEMANA**

```python
def semana(numero):
  dia = {
      1: "Lunes",
      2: "Martes",
      3: "Miércoles",
      4: "Jueves",
      5: "Viernes",
      6: "Sábado",
      7: "Domingo"
  }

  if numero in dia:
    return dia[numero]
  else:
    return "No existe"

numero=int(input("Escriba un número del 1 al 7: "))
print(semana(numero))
```
