**CONVERSOR DE DIAS DE LA SEMANA**

```python
numero=int(input("Escriba un número del 1 al 7:"))
if numero==1:
  print("Lunes")
elif numero==2:
  print("Martes")
elif numero==3:
  print("Miércoles")
elif numero==4:
  print("Jueves")
elif numero==5:
  print("Viernes")
elif numero==6:
  print("Sábado")
elif numero==7:
  print("Domingo")
else:
  print("No existe")
```

**CONVERSOR DE NOTAS**

```python
numero1=int(input("Escriba un número del 1 al 10:"))
if numero1<5:
  print("Insuficiente")
elif numero1>5 and numero1<6:
  print("Suficiente")
elif numero1>6 and numero1<7:
  print("Bien")
elif numero1>7 and numero1<9:
  print("Notable")
elif numero1>9 and numero1<10:
  print("Sobresaliente")
elif numero1==10:
  print("matricula de honor")
else:
  print("Esa nota no existe")
```

**CAJERO AUTOMÁTICO**

```python
saldo=5000
print("Elija una de las opciones por favor:")
print("1.Consulta saldo")
print("2.Retirar dinero")
print("3.Ingresar dinero")
numero=int(input())
if numero==1:
  print("Su saldo actual es de:",saldo,"€")
elif numero==2:
  retirar=int(input("¿Cuánto dinero desea retirar?"))
  saldo -= retirar
  print("Le queda un total de:",saldo,"€")
elif numero==3:
  ingresar=int(input("¿Cuánto dinero desea ingresar?"))
  saldo += ingresar
  print("Dispone de un total de:",saldo,"€")
else:
  print("Error!! Operación debe ser 1,2, ó 3")
```

**CONVERSOR DE FECHA**

```python
semana={1:"Lunes",
        2:"Martes",
        3:"Miércoles",
        4:"Jueves",
        5:"Viernes",
        6:"Sábado",
        7:"Domingo"}
mes_dic={1:"Enero",
         2:"Febrero",
         3:"Marzo",
         4:"Abril",
         5:"Mayo",
         6:"Junio",
         7:"Julio",
         8:"Agosto",
         9:"Septiembre",
         10:"Octubre",
         11:"Noviembre",
         12:"Diciembre"}
diasemana=int(input("Introduce el primer número, día de la semana:"))
diames=int(input("Introduce el segundo número, día del mes:"))
mes=int(input("Introduce el tercer número, mes del año:"))
año=int(input("Introduce el cuarto número, el año:"))

print("Hoy es" + semana[diasemana] + "," +str (diames)+ "de" + mes_dic[mes] + "de" +str (año))

```
